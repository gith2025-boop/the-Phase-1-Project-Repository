Analysis

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use('ggplot')
plt.boxplot(df['production_year'])
plt.show()
```

Most Common Genres

Most common genres that have been in production are Drama, Documentary and Comedy. other genres seem to be doing fairy similar in the market as per the graph below

```python
genres = df_subset2['genres'].value_counts().head(20).sort_values(ascending=True)

plt.figure(figsize=(20, 6))
genres.plot(kind='barh')
plt.title('Movie Genres')
plt.xlabel('Number of movies')
plt.show()
```

```python
#1 Which genres had most movies produced
df_subset2['genres'].value_counts().plot(kind='bar', title='Number of Items per genre')
```

Production vs Genres

Below is a scatter plot representing genres of movies produced verses their production years. from the scatter plot, we can also see the various common genres being common in the most recent years being Documentary History, Action and Drama, Comedy,  Documentary and Drama

```python
#3 which genres were produced ,and in which years
df_subset2.plot.scatter(x='production_year', y='genres', title='Production Year vs Genres', color='green', alpha=0.6, figsize=(8,6))
```
